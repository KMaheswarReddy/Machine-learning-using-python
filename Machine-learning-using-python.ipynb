```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


# code for url method- dont execute
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
print(df.head())

df = sns.load_dataset('iris')
print(df)

# Separating features and target

feature = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
x = df.loc[:,feature]  # x is feature/ data
y = df.loc[:,'species'] # y is labels/sample output
print(x)
print(y)


# Standardising features

s=StandardScaler() # initializing standard scaler
x=s.fit_transform(x) # stsclr.fit_transform
print(x)


# Creating the principal component

pca = PCA(n_components=2) # initialize PCA with no.of components
pct = pca.fit_transform(x)

print(pct)


principal_df = pd.DataFrame(pct,columns=['pc1','pc2'])
finaldf= pd.concat([principal_df,df[['species']]],axis=1)
print(finaldf)


# Plot the Principal component-1 and 2 using scater plot

ax =sns.scatterplot(x="pc1", y="pc2", hue="species",data=finaldf)

```
